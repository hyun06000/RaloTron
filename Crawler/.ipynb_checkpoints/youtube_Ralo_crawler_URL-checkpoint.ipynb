```python
channel_URL = "https://www.youtube.com/channel/UCD2YO_A_PVMgMDN9jpRrpVA/videos"
```

```python
req = requests.get(channel_URL)
req.headers['content-type']
```

```python
html = req.text
```

```python
soup = BeautifulSoup(html, "lxml")
```

```python
soup.encode()
```

```python
soup.find_all("a",{"href":"/watch?v=GIPfEWafstI"})
```

```python
my_titles = soup.select('h3 > a')
```

```python
url = []
for idx in my_titles:
    print(idx.get('href'))
    if idx.get('href')[:7] != '/watch?':
        continue
    else:
        url.append(idx.get('href'))
```
